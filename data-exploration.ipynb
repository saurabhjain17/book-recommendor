```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd
data=pd.read_csv(f"{path}\\books.csv")
```

```python
data
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
ax=plt.axes()
sns.heatmap(data.isna().transpose(),cbar=False,ax=ax)
plt.xlabel("columns")
plt.ylabel("Missing values")
```

```python
import numpy as np
data["missing_description"]=np.where(data["description"].isna(),1,0)
data["age_of_book"]=2025-data["published_year"]
```

```python
columns_of_interest=["num_pages","age_of_book","missing_description","average_rating"]
correlation_matrix=data[columns_of_interest].corr(method="spearman")
sns.set_theme(style="white")
plt.figure(figsize=(8,6))
heatmap=sns.heatmap(correlation_matrix,annot=True,fmt=".2f",cmap="coolwarm",cbar_kws={"label":"Spearman correlation"})
heatmap.set_title("correlation matrix")
plt.show()
```

```python
book_missing=data[~(data["description"].isna())&
                  
               ~(data["num_pages"].isna())&
                   ~(data["average_rating"].isna())&
                   ~(data["published_year"].isna())
                  ]
```

```python
book_missing
```

```python
book_missing["categories"].value_counts().reset_index().sort_values("count",ascending=False)
```

```python
book_missing["words_in_description"]=book_missing["description"].str.split().str.len()
```

```python
book_missing
```

```python
book_missing_25_words=book_missing.loc[book_missing["words_in_description"]>=25]
```

```python
book_missing_25_words
```

```python
book_missing_25_words["title_and_subtitle"]=(
    np.where(book_missing_25_words["subtitle"].isna(),book_missing_25_words["title"],
             book_missing_25_words[["title","subtitle"]].astype(str).agg(": ".join,axis=1))
)
```

```python
book_missing_25_words
```

```python
book_missing_25_words["tagged_description"]=book_missing_25_words[["isbn13","description"]].astype(str).agg(" ".join,axis=1)
```

```python
book_missing_25_words
```

```python
(
    book_missing_25_words.drop(["subtitle","missing_description","age_of_book","words_in_description"],axis=1).to_csv("books_cleaned.csv",index=False)
)
```
